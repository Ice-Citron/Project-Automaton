The task: train an AI to autonomously INSERT FIBER OPTIC CABLES (SFP, LC, SC
connectors) into a simulated electronics assembly workstation. This mirrors
real-world data center and server manufacturing where cable management is still
done manually because it's so hard to automate -- cables are flexible, 
connectors require extreme precision, and the workspace is dense.

-- ... file extensions (.glb, .xacro and .sdf) are commonly used in 3D modelling
   , simulation, and robotics (specifically with ROS - Robot Operating System
   and Gazebo) to define 3D assets, robot structures, and simulation 
   environments.

   1. `.glb` (GL TRANSMISSION FORMAT BINARY)
      - WHAT IT IS: The binary file format version of glTF (GL Transmission
        Format). It is often referred to as the "JPEG of 3D".
      - PURPOSE: It stores 3D model information--including geometry, textures,
        materials, and animation--in a single, self-contained binary file.
      - USAGE: Used for sharing 3D models efficiently in web apps, augmented 
        reality (AR), virtual reality (VR), and simulation environments (like
        Gazebo/Ignition).
      - KEY FEATURE: Because it is binary, it is faster to load and smaller in
        size compared to text-based formats (like .obj or .dae).

   2. `.xacro` (XML Macros)
      - WHAT IT IS: An extension of the Unified Robot Description Format (URDF)
        used in ROS (Robot Operating System)
      - PURPOSE: It allows you to use macros, parameterise models, and include
        other files to make robot descriptions shorter, more readable, and
        easier to maintain.
      - USAGE: A `.xacro` file is processed (compiled) to produce a standard 
        `.urdf` or `.sdf` file.
      - EXAMPLE: Instead of copying and pasting the code for a 4-legged robot,
        you define one "leg" macro and call it four times.

   3. `.sdf` (Simulation Description Format)
      - WHAT IT IS: An XML-based file format used primarily by the Gazebo 
        physics simulator to describe everything 
      - PURPOSE: It provides a comprehensive description of robots, lightining,
        terrain, and physics properties.
      - USAGE: While URDF is used for modelling robots in ROS (for kinematics),
        SDF is generally preferred for full-scene simulation in Gazebo because
