   Rather than forcing you to manually start simulations, launch ROS2 nodes, and

   robotic policy you are evaluating.

   Here is a breakdown of the core concepts of what this script is doing 
   relative to the `intrinsic-dev/aic` repository.


### 1. The Orchestrator (Docker Compose Wrapper)
   At its heart, this script is a sophisticated wrapper around Docker Compose.
   THe `intrinsic-dev/aic` repository relies heavily on containerisation to
   ensure the Gazebo physics simulation and the evaluation engine run 
   consistently regardless of your host machine.

   - WHAT IT DOES: It programatically builds and executes `docker compose up`
     commands.
   - WHY IT MATTERS: It manages the lifecycle of the simulation containers, 
     parses their logs in real-time (`stream_subprocess_output`), and ensures
     everything is cleanly torn down when the evaluation finishes or if it times
     out (`teardown_compose`).  


### 2. THE TWO EXECUTION MODES (Local vs. Dockerised Policy)
   The script supports two distinctly different ways to test your robotic
   policy against the simulation:
      - LOCAL POLICY (The Default): The script starts the `eval` container
        (Gazebo simulation) in the background. It then waits for a TCP port
        (7447) to open, which indicates Zenoh (the communication protocol)

-- Docker Compose is a tool for defining and running multi-container 
   applications. It is the key to unlocking a streamlined and efficient 
   development and deployment experience. 

-- Zenoh is an open-source, high-performance, and extremely low-overhead 
   protocol that unifies data in motion (pub/sub), data at rest (storage), and
   computations. Designed... it spans environments from resource-constrained
   micro-controllers (300-byte footprint) to data centers.   

   Key features and benefits of Zenoh include:
   * HIGH PERFORMANCE: Delivers over 4 million messages per second with ms-level
     latency (15 microseconds), enabling efficient communication for robotics,
     IoT, and V2X (Vehicle-to-Anything) applications.
   * UNIFICATION: It bridges the gap between pub/sub, distributed storage, and 
     query mechanisms, allowing applications to seamlessy access data regardless
     of whether it is currently in motion or stored.
   * LOW OVERHEAD / RESOURCE EFFICIENCY: Features a minimal 4-6 byte wire 
     overhead and operates efficiently on low-power networks (LPWAN) and
     constrained devices.
   * FLEXIBLE TOPOLOGIES: Support peer-to-peer, routed, and brokered 
     communication models, allowing it to work on top of TCP/IP, UDP/IP, and
     diectly over Ethernet or serial links.
     ...
   
   Zenoh is designed for systems where low latency and high throughput are 
   critical, such as robotics and smart city infrastructure.
