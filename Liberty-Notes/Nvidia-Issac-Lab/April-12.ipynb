                The short answer to your question is: VQ-GAN does not directly accept arbitrary image sizes as-is. During tokenizer
                training, images are first standardized by resize/crop/normalize, then an encoder maps them into a smaller continuous
                latent grid, and then each latent vector is snapped to its nearest learned codebook vector. The codebook is not
                precomputed like BPE merges in text tokenizers; it is learned jointly with the encoder and decoder using reconstruction,
                perceptual, adversarial, codebook, and commitment losses. So it is similar to a tokenizer in that it creates a discrete
                vocabulary before transformer pretraining, but the mechanism is learned vector quantization rather than corpus-frequency
                merge rules.
