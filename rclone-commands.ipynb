## Part 1: Set up rclone with Google Drive on Vast.ai

**1. Install rclone:**
```bash
apt install rclone -y
```

**2. Start configuration:**
```bash
rclone config
```

**3. Follow the prompts:**
```
n                          # new remote
gdrive                     # name it
17                         # Google Drive (number may vary, look for "drive")
<Enter>                    # leave client_id blank
<Enter>                    # leave client_secret blank
1                          # full access
<Enter>                    # leave root_folder_id blank
<Enter>                    # leave service_account_file blank
n                          # no advanced config
n                          # no auto config (headless server)
```

**4. It will give you a URL.** Open it on your Mac/Windows browser, log into the Google account you want, authorize, and copy the code back into the terminal.

**5. Confirm and quit:**
```
y                          # yes this is ok
q                          # quit config
```

---

## Part 2: Sync /workspace to a specific Google Drive folder

**Create destination folder and sync:**
```bash
# Create folder on Google Drive
rclone mkdir gdrive:VastAI-Isaac-Sim

# Sync /workspace to that folder
rclone sync /workspace gdrive:VastAI-Isaac-Sim --progress
```

**Or for ongoing backup (copy without deleting):**
```bash
rclone copy /workspace gdrive:VastAI-Isaac-Sim --progress
```

**To sync a subfolder only:**
```bash
rclone sync /workspace/my-project gdrive:VastAI-Isaac-Sim/my-project --progress
```

## Part 3: Download from Google Drive to your Mac

**1. Install rclone on Mac:**
```bash
brew install rclone
```

**2. Configure (same process):**
```bash
rclone config
```

This time you can use auto config since you have a browser:
```
n
gdrive
17
<Enter>
<Enter>
1
<Enter>
<Enter>
n
y                          # YES to auto config (you have a browser)
```

Browser will open automatically â€” authorize and done.

**3. Download to Mac:**
```bash
# Download entire VastAI backup
rclone copy gdrive:VastAI-Isaac-Sim ~/Documents/VastAI-Isaac-Sim --progress

# Or sync (mirrors exactly, deletes local files not on Drive)
rclone sync gdrive:VastAI-Isaac-Sim ~/Documents/VastAI-Isaac-Sim --progress
```

## Quick reference commands

| Task | Command |
|------|---------|
| List Drive contents | `rclone ls gdrive:` |
| List folders only | `rclone lsd gdrive:` |
| Upload /workspace | `rclone copy /workspace gdrive:VastAI-Isaac-Sim -P` |
| Download to Mac | `rclone copy gdrive:VastAI-Isaac-Sim ~/local-folder -P` |
| Check what would sync | `rclone sync /workspace gdrive:VastAI-Isaac-Sim --dry-run` |

The `-P` is shorthand for `--progress`.
